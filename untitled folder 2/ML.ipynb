We are multiplying:

- **Matrix A**: size \( n , k \)  
- **Matrix B**: size \( k , m \)  

The result will be **Matrix C**: size \( n , m \).

---

### **How each element is computed**
Each element \($ C_{ij}$ \) is calculated as:
$$ C_{ij} = \sum_{t=1}^{k} A_{it} \cdot B_{tj} $$



This means:
- **Multiplications per element**: k 
- **Additions per element**: k - 1 

---

### **Total operations**
Since there are ($n \times m$) elements in the result:

$$
\text{Total multiplications} = n \times m \times k
$$

$$
\text{Total additions} = n \times m \times (k - 1)
$$

---

**Final Answer:**
- Multiplications: \( $n \times m \times k $\)  
- Additions: \( $n \times m \times (k - 1) $\)

### QUESTION 3
#### Finding the highest element in a list requires one pass of the array. Finding the second highest element requires 2 passes of the array. Using this method, what is the time complexity of finding the median of the array? Can you suggest a better method? Can you implement both these methods in Python and compare against numpy.median routine in terms of time?

Highest element -> 1 pass -> O(n)
Second highest element -> 1 pass -> O(n)

- **Median** using the same “pass for each order statistic” method:
  - For the median (middle element), you’d need **n/2 passes** in the worst case.
  - Each pass is **O(n)**, so total is:

$
O\left(\frac{n}{2} \times n\right) = O(n^2)
$

- **Conclusion:** This method is **very inefficient** for large arrays.


```python
n = 10000
lst = [random.randint(1, 1000000) for _ in range(n)]



def median_repeated_passes(arr):
    temp_lst = arr.copy()
    target_index = len(temp_lst) // 2
    for _ in range(target_index):
        max_val = max(temp_lst)
        temp_lst.remove(max_val)
    return max(temp_lst)

def median_sort(arr):
    sorted_lst = sorted(arr)
    mid = len(sorted_lst) // 2
    if len(sorted_lst) % 2 == 1:
        return sorted_lst[mid]
    else:
        return (sorted_lst[mid - 1] + sorted_lst[mid]) / 2


def median_numpy(arr):
    return np.median(arr)

start = time.time()
median1 = median_repeated_passes(lst)
time1 = time.time() - start

start = time.time()
median2 = median_sort(lst)
time2 = time.time() - start

start = time.time()
median3 = median_numpy(lst)
time3 = time.time() - start

print(f"Repeated passes median: {median1} | Time: {time1:.6f} sec")
print(f"Sort-based median:      {median2} | Time: {time2:.6f} sec")
print(f"NumPy median:           {median3} | Time: {time3:.6f} sec")
    
```

### Question 4
#### What is the gradient of the following function with respect to x and y?

We have:
We have:

$$
f(x, y) = x^2 y + y^3 \sin(x)
$$

---

### **Step 1: Partial derivative with respect to $x$**

- For $x^2 y$, treat $y$ as constant: derivative is $2xy$  
- For $y^3 \sin(x)$, $y^3$ is constant w.r.t. $x$: derivative is $y^3 \cos(x)$  

$$
\frac{\partial f}{\partial x} = 2xy + y^3 \cos(x)
$$

---

### **Step 2: Partial derivative with respect to $y$**

- For $x^2 y$, derivative w.r.t. $y$ is $x^2$  
- For $y^3 \sin(x)$, derivative w.r.t. $y$ is $3y^2 \sin(x)$  

$$
\frac{\partial f}{\partial y} = x^2 + 3y^2 \sin(x)
$$

---

### **Final Gradient**

$$
\nabla f(x, y) = \left[\, 2xy + y^3 \cos(x), \; x^2 + 3y^2 \sin(x) \,\right]
$$

### Question 5
#### Use JAX to confirm the gradient evaluated by your method matches the analytical solution corresponding to a few random values of x and y


```python
import jax
import jax.numpy as jnp

def f(x, y):

    return x**2 * y + y**3 * jnp.sin(x)

grad_f = jax.grad(f, argnums=(0, 1))

def analytical_grad(x, y):
    gx = 2 * x * y + y**3 * jnp.cos(x)       
    gy = x**2 + 3 * y**2 * jnp.sin(x)       
    return gx, gy


rng = np.random.default_rng(42)
num_tests = 6
tol = 1e-6

print("Comparing analytical gradient to JAX autodiff gradient\n")

for i in range(num_tests):
    
    x0 = rng.uniform(-5.0, 5.0)
    y0 = rng.uniform(-5.0, 5.0)


    xj = jnp.array(x0)
    yj = jnp.array(y0)


    jax_gx, jax_gy = grad_f(xj, yj)
    anal_gx, anal_gy = analytical_grad(xj, yj)


    jax_gx_n = float(jax_gx)
    jax_gy_n = float(jax_gy)
    anal_gx_n = float(anal_gx)
    anal_gy_n = float(anal_gy)
    print(f"Test {i+1}: x={x0:.6f}, y={y0:.6f}")
    print(f"  JAX grad   -> (df/dx, df/dy) = ({jax_gx_n:.9f}, {jax_gy_n:.9f})")
    print(f"  Analytic   -> (df/dx, df/dy) = ({anal_gx_n:.9f}, {anal_gy_n:.9f})")

    # check close
    ok_x = np.isclose(jax_gx_n, anal_gx_n, atol=tol, rtol=1e-6)
    ok_y = np.isclose(jax_gy_n, anal_gy_n, atol=tol, rtol=1e-6)
    if ok_x and ok_y:
        print("Match within tolerance\n")
    else:
        print("Mismatch!")
        print(f"    diff dx = {jax_gx_n - anal_gx_n:.9e}")
        print(f"    diff dy = {jax_gy_n - anal_gy_n:.9e}\n")
```

```python
def my_argmin(arr):

    min_index = 0
    min_value = arr[0]
    

    for i in range(1, len(arr)):
        if arr[i] < min_value:
            min_value = arr[i]
            min_index = i
    
    return min_index


arr = np.array([3, 1, 4, 1.5, -2, 7])


my_index = my_argmin(arr)


np_index = np.argmin(arr)

print("Array:", arr)
print("My argmin index:", my_index)
print("NumPy argmin index:", np_index)
```
